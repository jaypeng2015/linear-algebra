## Example 1

Linear equations:

\begin{equation*} 2x-y=0 \end{equation*}
\begin{equation*} -x+2y=3 \end{equation*}

Matrix form:

\begin{equation*}
\begin{vmatrix}2 & -1 \\ -1 & 2\\\end{vmatrix}
\begin{vmatrix}x\\ y\\ \end{vmatrix} = 
\begin{vmatrix}0\\ 3\\ \end{vmatrix}
\end{equation*}
\begin{equation*} Ax=b \end{equation*}

```python
# The row picture
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
fig = plt.figure()
ax = plt.axes()

x = np.linspace(-10, 10)
ax.plot(x, np.multiply(x, 2));
ax.plot(x, np.divide(np.add(x, 3), 2));
```

### Linear Combination

The most fundamental operation of the whole course. Also this is the column picture.

\begin{equation} x\begin{vmatrix}2 \\ -1\end{vmatrix} + y\begin{vmatrix}-1 \\ 2\end{vmatrix} = 
\begin{vmatrix}0 \\ 3\end{vmatrix}
\end{equation}

It's the linear combination of the columns of A.

```python
import matplotlib.pyplot as plt
import numpy as np


V = np.array([[2,-1],[-1,2],[0,3]])
origin = [0], [0] # origin point

plt.quiver(*origin, V[:,0], V[:,1], color=['r','b','g'], angles='xy', scale_units='xy', scale=1)
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.show()
```

## Example 2

\begin{equation*} 2x-y=0 \end{equation*}
\begin{equation*} -x+2y-z=-1 \end{equation*}
\begin{equation*} -3y+4z=4 \end{equation*}

\begin{equation*}
\begin{vmatrix}2 & -1 & 0\\ -1 & 2 & -1\\0 & -3 & 4\\\end{vmatrix}
\begin{vmatrix}x\\ y\\ z\\ \end{vmatrix} = 
\begin{vmatrix}0\\ -1\\ 4 \\ \end{vmatrix}
\end{equation*}

```python
# The row picture
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# equation 1
x = np.linspace(-10, 10)
z = np.linspace(-10, 10)
x, z = np.meshgrid(x, z)
y=np.multiply(x, 2)
ax.plot_surface(x, y, z)

# equation 2
x = np.linspace(-10, 10)
y = np.linspace(-10, 10)
x, y = np.meshgrid(x, y)
z=np.subtract(2*y, (x+1))
ax.plot_surface(x, y, z)

# equation 3
x = np.linspace(-10, 10)
y = np.linspace(-10, 10)
x, y = np.meshgrid(x, y)
z=np.divide(3*y+4, 4)
ax.plot_surface(x, y, z)


```

The linear combination form:

\begin{equation} x\begin{vmatrix}2 \\ -1 \\ 0\end{vmatrix} + y\begin{vmatrix}-1 \\ 2 \\ -3\end{vmatrix} + z\begin{vmatrix}0 \\ -1 \\ 4\end{vmatrix}= 
\begin{vmatrix}0 \\ -1 \\ 4\end{vmatrix}
\end{equation}

So the solution is x=0,y=0,z=1.

## Question

Can I solve Ax=b for every b?

In other words, can Ax cover the whole space?

The answer for the A in example 2 is yes. But generely speaking, if one of the columns is the combination of some of the other columns, then the answer will be no.

## Different ways of Matrix Mutiplication

### option 1, by columns

\begin{equation}
\begin{vmatrix}2 & 5 \\ 1&3\end{vmatrix}\begin{vmatrix}1 \\ 2\end{vmatrix}
= 1\begin{vmatrix}2 \\ 1\end{vmatrix} + 2 \begin{vmatrix}5 \\ 3\end{vmatrix}
= \begin{vmatrix}12 \\ 7\end{vmatrix}
\end{equation}

### option 2, by rows
\begin{equation*}
\begin{vmatrix}2 & 5 \\ 1&3\end{vmatrix}\begin{vmatrix}1 \\ 2\end{vmatrix}
= \begin{vmatrix}{2 \times 1 + 5 \times 2} \\ {1 \times 1 + 3 \times 2} \end{vmatrix}
= \begin{vmatrix}12 \\ 7\end{vmatrix}
\end{equation*}

By columns is easier to understand: Ax is the combination of columns of A.
