L: a line through the origin but not in P

Is (P ∪ L) a subspace? No.

Is (P ∩ L) a subspace? Yes. It's the origin in this case.

## Review the rules of a vector space

 - v + w and cv are in the space
 - all cv + dw are in the space

## Column Space of Matrix A - C(A)

\begin{equation}
A=\begin{vmatrix}1&1&2\\2&1&3\\3&1&4\\4&1&5\end{vmatrix}
\end{equation}

The column space is in R<sup>4</sup> because the vectors have 4 components.

The column space is all the linear combination of the columns.

So if we have a system:

\begin{equation}
Ax=\begin{vmatrix}1&1&2\\2&1&3\\3&1&4\\4&1&5\end{vmatrix}\begin{vmatrix}x_1\\x_2\\x_3\end{vmatrix}=b
\end{equation}

Does it alway have solutions for every b?

Not really because there are 3 unkowns but4 equations.

Based on the definition, b needs to be the combinations of columns in A, in other words, b needs be in the column space of A.

"I can solve Ax=b exactly when b is a vector in the column space of A."

The this example column 3 contributes nothing new, and the column space is a 2 dimenional plane in R<sup>4</sup>.

## Null Space of Matrix A  - N(A)

\begin{equation}
Ax=\begin{vmatrix}1&1&2\\2&1&3\\3&1&4\\4&1&5\end{vmatrix}\begin{vmatrix}x_1\\x_2\\x_3\end{vmatrix}=\begin{vmatrix}0\\0\\0\\0\end{vmatrix}
\end{equation}

The null space is the solution space of the above system.

x have 3 components, so the solution space is in R<sup>3</sup>.

The zero vector is obviously one solution.

All the solutions:

\begin{align}
c\begin{vmatrix}1\\1\\-1\end{vmatrix}
\end{align}

And this is actually a vector space, a line in R<sup>3</sup>.

```python
## The special solution
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

soa = np.array([[0, 0, 0, 1, 1, -1]])

X, Y, Z, U, V, W = zip(*soa)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W)
ax.set_xlim([-2, 2])
ax.set_ylim([-2, 2])
ax.set_zlim([-2, 2])
plt.show()
```

```python
# The whole subspace
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.gca(projection='3d')
x1 = np.linspace(-100, 100, 100)
x2 = x1
x3 = -x1
ax.plot(x1, x2, x3, label='the null space')
ax.legend()
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('x3')
```
